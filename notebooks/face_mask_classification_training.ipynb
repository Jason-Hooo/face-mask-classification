## 5. Model Training & Optimization

Initiate the training process using a Transfer Learning approach with the **EfficientNet** backbone. The `train.py` script is designed with a flexible CLI interface for hyperparameter tuning and reproducibility.

### CLI Arguments & Technical Specifications:
- `--model_name`: EfficientNet variant selection (Choices: `B0`-`B7`). Currently set to `B1` for optimal balance.
- `--epochs`: Total training iterations over the dataset (Default: `25`).
- `--batch_size`: Number of training examples utilized in one iteration (Default: `32`).
- `--learning_rate`: Step size for the optimizer (Default: `1e-3`).
- `--dataset_root`: Path to the processed dataset directory (Default: `Face_Mask_Dataset`).
- `--save_path`: Destination path for the serialized model weights (`.pth`).
- `--seed`: Global random seed to ensure deterministic behavior and reproducibility (Default: `87`).

```python
!PYTHONPATH=src python src/train.py --epochs 25 --batch_size 32 --model_name B1
```
